```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_excel("/content/Solar Dataset_Solar from MEDA-1.XLSX")
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df=df[[ '#date time', 'el','az','ghi','wb','precip','press','dp','rh','dhi_flag','dni','ws','maxws','airtemp']]
```

```python
import pandas as pd


# Extract features (year, month, day, hour, minute, weekday)
df['year'] = df['#date time'].dt.year
df['month'] = df['#date time'].dt.month
df['day'] = df['#date time'].dt.day
df['hour'] = df['#date time'].dt.hour
df['minute'] = df['#date time'].dt.minute
df['weekday'] = df['#date time'].dt.weekday
```

```python
def print_unique_col_values(df):
       for column in df:
            if df[column].dtypes=='object':
                print(f'{column}: {df[column].unique()}')
```

```python
df.replace(' NaN',0,inplace=True)
```

```python
print_unique_col_values(df)
```

```python
print(df.isnull().sum())
print(df.isin([np.nan, np.inf, -np.inf]).sum())


# Replace NaN and infinity values with zeros
df.replace([np.nan, np.inf, -np.inf], 0, inplace=True)
```

```python
df=df.drop('#date time', axis=1)
```

```python
df=df.drop('dhi_flag', axis =1)
df=df.drop('year',axis =1)
```

```python
#df=df.drop('dni',axis =1)
```

```python
df
```

```python
# Writing ARIMA model to predict 'dni'

from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error

# Split the data into training and test sets
train_data = df[:27700]
test_data = df[27700:]

# Fit the ARIMA model
model = ARIMA(train_data['dni'], order=(5, 1, 0))
model_fit = model.fit()

# Make predictions on the test set
predictions = model_fit.forecast(len(test_data))

# Calculate the mean squared error
error = mean_squared_error(test_data['dni'], predictions)
print('Mean squared error:', error)
```

```python
# applying sarima model to predict dni

from statsmodels.tsa.statespace.sarimax import SARIMAX

# Fit the SARIMA model
model = SARIMAX(train_data['dni'], order=(5, 1, 0), seasonal_order=(1, 0, 1, 12))
model_fit = model.fit()

# Make predictions on the test set
predictions = model_fit.forecast(len(test_data))

# Calculate the mean squared error
error = mean_squared_error(test_data['dni'], predictions)
print('Mean squared error:', error)
```

```python
# applying fb prophet to predict dni

from prophet import Prophet

# Create a Prophet model
model = Prophet()

# Add the DNI column as the y variable
model.add_regressor('dni')

# Fit the model on the training data
model.fit(train_data)

# Make predictions on the test set
forecast = model.predict(test_data)

# Calculate the mean squared error
error = mean_squared_error(test_data['dni'], forecast['yhat'])
print('Mean squared error:', error)
```
